```python
import pandas as pd
df=pd.read_csv('/content/nyc_weather.csv')
df.head()
```

```python
max_temperature=df['Temperature'].max()
```

```python
rainy_days=df[df['Events'].str.contains('Rain',na=False)]
rainy_dates=rainy_days['EST']
print(rainy_dates)
```

```python
average_wind_speed=df['WindSpeedMPH'].mean()
```

```python
print ("Maximum temperature in New York(January):",max_temperature,"°F")
print("Days it rained:",rainy_days)
print("Average wind speed:",
      round(average_wind_speed,2),"mph")
```
